Importing dependencies


```python
#importing some required dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
#Data collection from pc 
sonar_data=pd.read_csv('/content/sonar_data.csv',header=None)
```

```python
#showing some data
sonar_data.head(20)
```

```python
# number of rows and columns
sonar_data.shape
```

```python
#showing details about dataset
sonar_data.describe()
```

```python
#details about outcomes
sonar_data[60].value_counts()
```

```python
#Grouped by R and M 
sonar_data.groupby(60).mean()
```

Sepating data and lebels

```python
X = sonar_data.drop(columns=60,axis=1)
Y = sonar_data[60]
```

```python
#showing X and Y
print(X)
print(Y)
```

Training and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training


```python
#model selection
model = LogisticRegression()
```

```python
#training the model with training data
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print("Accuracy on traing data",training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print("Accuracy on test data",test_data_accuracy)
```
